### Q2: 判斷基數偶數
<b>輸入說明<br></b>
輸入一個整數<br>
<b>輸出說明 <br></b>
輸出 odd或 even<br>
不合法的輸入則輸出E。<br>

<blockquote>
輸入範例:<br>
3<br>
輸出範例:<br>
odd<br>

輸入範例:<br>
6<br>
輸出範例:<br>
even<br>
</blockquote>

```python
x =float( input())
if (x%2 == 0):
    print("even")
if (x%2 == 1):
    print("odd")
else:
    print("E")
```

### Q3: 計算總成績、平均 
<b>說明：<br></b>
某一學生修國文、計算機概論、計算機程式設計三科，使用者輸入名字（一個char）、學號（int）、三科成績(int)。<br>
(1) 計算學生總成績、平均。<br>
(2) 印出名字、學號、總成績、平均。<br>
(3) 每個科目分數為0~100分。<br>

--------------------------------------------------------------------------------------------------------------------------- 
<b>輸出說明 <br></b>
輸出平均分數(整數)。若輸入格式有誤則輸出「輸入格式錯誤」。 <br>

--------------------------------------------------------------------------------------------------------------------------- 
<blockquote>
Input 1:<br>
Miso<br>
905067<br>
100<br>
100<br>
100<br>

Output 1:<br>
Name:Miso<br>
ID:905067<br>
Average:100<br>
Total:300<br>

Input 2: 若輸入不符合格式 即停止程式並輸出「輸入格式錯誤」<br>
Miso<br>
905067<br>
100<br>
a<br>

Output 2:<br>
輸入格式錯誤<br>
</blockquote>

```python
x = input("Name: ")
y = input("ID: ")
try:
    a,b,c =list(map(int,input().split()))
except:
    print("格式輸入錯誤")
else:
    z = int((a+b+c)/3)
    print("Name: ", x ,"\nID: ", y )
    print("Avg: ",z)
```
